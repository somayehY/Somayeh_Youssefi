```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from random import seed
```

```python
normal_distribution = np.random.normal(loc=5, scale=1, size=10000)
normal_distribution = pd.DataFrame(normal_distribution, columns=['value'])
uniform_distribution = np.random.uniform(size=10000)
uniform_distribution = pd.DataFrame(uniform_distribution, columns=['value'])
lognormal_distribution = np.random.lognormal(mean=0, sigma=0.5, size=10000)
lognormal_distribution= pd.DataFrame(lognormal_distribution, columns=['value'])
```

```python
datasets = [normal_distribution, uniform_distribution, lognormal_distribution]
titles = ['Normal Distribution', 'Uniform Distribution', 'Lognormal Distribution']
colors = ['skyblue', 'limegreen', 'orange']
```

```python
fig, [ax0, ax1, ax2] = plt.subplots(1,3, figsize=(18,4))
for ax, dataset, color, title in zip([ax0, ax1, ax2], datasets, colors, titles):
    sns.histplot(x='value', data=dataset, bins=25,color=color, ax=ax)
    ax.set_title(title)
```

```python
def sampling_averaging_function(dataset, repeat_size, sample_size):
    seed(47)
    sample_means = []
    for _ in range(0, repeat_size):
        sample = dataset['value'].sample(sample_size)
        sample_means.append(sample.mean())
    return sample_means
```

```python

fig, [(ax0, ax1, ax2), (ax3, ax4, ax5), (ax6, ax7, ax8), (ax9, ax10, ax11)] = plt.subplots(4,3, figsize=(18,16))
titles = ['25 Means Sampled from a Normal Distribution', \
        '25 Means Sampled from a Uniform Distribution', \
        '25 Means Sampled from a Lognormal Distribution']
colors = ['aquamarine', 'limegreen', 'lightsalmon']
for ax, dataset, color, title in zip([ax0, ax1, ax2], datasets, colors, titles):
    sample_means = sampling_averaging_function(dataset, 25, 50)
    sns.histplot(x=sample_means,  color=color, bins=25,ax=ax)
    ax.set_title(title)
titles = ['50 Means Sampled from a Normal Distribution', \
        '50 Means Sampled from a Uniform Distribution', \
        '50 Means Sampled from a Lognormal Distribution']
colors = ['lightseagreen', 'mediumseagreen', 'salmon']
for ax, dataset, color, title in zip([ax3, ax4, ax5], datasets, colors, titles):
    sample_means = sampling_averaging_function(dataset, 50, 50)
    sns.histplot(x=sample_means,  color=color, bins=25,ax=ax)
    ax.set_title(title)

titles = ['100 Means Sampled from a Normal Distribution', \
        '100 Means Sampled from a Uniform Distribution', \
        '100 Means Sampled from a Lognormal Distribution']
colors = ['royalblue', 'seagreen', 'indianred']
for ax, dataset, color, title in zip([ax6, ax7, ax8], datasets, colors, titles):
    sample_means = sampling_averaging_function(dataset, 100, 50)
    sns.histplot(x=sample_means,  color=color, bins=25,ax=ax)
    ax.set_title(title)

titles = ['1000 Means Sampled from a Normal Distribution', \
        '1000 Means Sampled from a Uniform Distribution', \
        '1000 Means Sampled from a Lognormal Distribution']
colors = ['navy', 'darkgreen', 'darkred']
for ax, dataset, color, title in zip([ax9, ax10, ax11], datasets, colors, titles):
    sample_means = sampling_averaging_function(dataset, 1000, 50)
    sns.histplot(x=sample_means,  color=color, bins=25,ax=ax)
    ax.set_title(title)
    
```
